## Rearrange headers

```python
"""
Function to spread list values left to right.
Used to deal with implict multiple headers.
["","value","","value2",""] ==> ["","value","value","value2","value2"]
"""
def spread_list(inlist):
    outlist = []
    curr = ""
    for item in inlist:
        if(not item.startswith('Unnamed')): # hit a new one, update
            curr = item.strip()
        outlist.append(curr) # append old or new item
    return outlist
```

```python
with open('tableau_total.csv') as csvfile:

  # Must use csv.reader to get each row as a list.
  myCSVReader = csv.reader(csvfile, delimiter=",", quotechar='"')

  header1 = spread_list(next(myCSVReader)) # first row is sectors
  header2 = spread_list(next(myCSVReader)) # second row is genders
  header3 = spread_list(next(myCSVReader)) # "% of male employment"
  header4 = spread_list(next(myCSVReader)) # periods.

print(len(header1))
print(len(header2))
print(len(header3))
print(len(header4))

header= [header1,header2,header3,header4]
```

```python
def drop_last_rows(df):
    nb_unique_values_per_row=df.nunique(axis=1)
    end_of_dataframe=nb_unique_values_per_row[nb_unique_values_per_row == 1].index.tolist()[0] #rÃ©cupÃ©rer l'index de la ligne ou il y'a une seul et unique valeur (END)
    return df[0:end_of_dataframe]
```

### Extract three first columns

```python
def concatenate_id_and_species(id_and_raw_data, species_df):
    id_and_species=pd.concat([id_and_raw_data, species_df], axis=1)
    id_and_species.columns = id_and_species.columns.droplevel(1)
    return id_and_species
```

### Extract planted species

```python
def get_planted_species(df):
    index_begin_plantation=dataframe_total_cleaned.columns.get_loc(('PLANTATION', 'Plantation or Delivery date', "Translation", "Date_of_Planting(1)"))
    index_end_plantation=dataframe_total_cleaned.columns.get_loc(('PLANTATION', 'Total planted trees', "Translation", ""))
    df_plantation = dataframe_total_cleaned.iloc[:, index_begin_plantation+1:index_end_plantation]
    df_plantation.columns = df_plantation.columns.droplevel(0)
    return df_plantation
```

```python
planted_species=get_planted_species(dataframe_total_cleaned)
planted_species.head()
```

```python
planted_species_with_id=concatenate_id_and_species(first_three_columns,planted_species)
planted_species_with_id.head()
```

```python
def get_m1_species(df):
    index_begin_m1=dataframe_total_cleaned.columns.get_loc(('Monitoring 1', 'Technician name', "Translation", "Technician_name_parcel(1)"))
    index_end_m1=dataframe_total_cleaned.columns.get_loc(('Monitoring 1', 'Total replanted trees at mon 1', "Translation", ""))
    df_m1_species = dataframe_total_cleaned.iloc[:, index_begin_m1+1:index_end_m1]
    df_m1_species.columns = df_m1_species.columns.droplevel(0) # get every species from monitoring m1

    df_replanted_m1 = df_m1_species.iloc[:, :int(len(df_m1_species.columns)/2)] # get replanted m1 species by splitting the dataframe by half
    df_living_m1 = df_m1_species.iloc[:, int(len(df_m1_species.columns)/2):]
    return df_replanted_m1,df_living_m1
```

## Write to excel

```python
# https://stackoverflow.com/questions/57961010/how-to-hide-the-rows-index/57961084#57961084

def write_excel(writer, df, start, *args, **kwargs):
    """
    Write df as an excel file to path, roughly similar to `df.to_excel` except that it handles
    `df` with MultiIndex columns and `index=False`.
    """

    header = pd.DataFrame(df.columns.to_list()).T
    header.to_excel(writer, startrow=start, header=False, *args, **kwargs)

    # Avoid the "NotImplementedError: Writing to Excel with MultiIndex columns"
    # exception by temporarily changing the columns to a single-level index
    orig_columns = df.columns
    df.columns = range(len(df.columns))
    df.to_excel(
        writer, startrow=len(header)+start, header=False, *args, **kwargs
    )
    df.columns = orig_columns
    writer.save()
```

```python
output_file=r"File_with_splitted_sheets.xlsx"

with pd.ExcelWriter(output_file) as writer:
    write_excel(writer, planted_species_with_id, output_file, 2, sheet_name="Colonne_Plantation", index=False) 
    write_excel(writer, replanted_m1_with_id, output_file, 2, sheet_name="Niveau1_M1", index=False) 
    write_excel(writer, living_m1_with_id, output_file, 2, sheet_name="Niveau2_M1", index=False) 
    write_excel(writer, replanted_m2_with_id, output_file, 2, sheet_name="Niveau1_M2", index=False) 
    write_excel(writer, living_m2_with_id, output_file, 2, sheet_name="Niveau2_M2", index=False) 
```
