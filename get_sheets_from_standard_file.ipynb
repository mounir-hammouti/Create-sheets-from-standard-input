## Rearrange headers

```python
"""
Function to spread list values left to right.
Used to deal with implict multiple headers.
["","value","","value2",""] ==> ["","value","value","value2","value2"]
"""
def spread_list(inlist):
    outlist = []
    curr = ""
    for item in inlist:
        if(not item.startswith('Unnamed')): # hit a new one, update
            curr = item.strip()
        outlist.append(curr) # append old or new item
    return outlist
```

```python
with open('tableau_total.csv') as csvfile:

  # Must use csv.reader to get each row as a list.
  myCSVReader = csv.reader(csvfile, delimiter=",", quotechar='"')

  header1 = spread_list(next(myCSVReader))
  header2 = spread_list(next(myCSVReader))
  header3 = spread_list(next(myCSVReader))
  header4 = spread_list(next(myCSVReader))

print(len(header1))
print(len(header2))
print(len(header3))
print(len(header4))

header= [header1,header2,header3,header4]
```

### Clean last rows

```python
def drop_last_rows(df):
    df_with_no_end_delimiter=df
    nb_unique_values_per_row=df.nunique(axis=1)
    rows_with_single_value=nb_unique_values_per_row[nb_unique_values_per_row == 1].index.tolist() #récupérer les index des lignes ou il y'a une seul et unique valeur (END)
    if (len(rows_with_single_value) != 0):
        end_of_dataframe=rows_with_single_value[0] # récupérer la premiere qui ne contient qu'une seule valeur (END)
        df_with_no_end_delimiter = df[0:end_of_dataframe]
    return df_with_no_end_delimiter
```

## Extract Species

### Extract three first columns

```python
first_three_columns=dataframe_total_cleaned.iloc[:, : 3]
first_three_columns.columns = first_three_columns.columns.droplevel(0) # drop level with no header
first_three_columns.head()
```

```python
def concatenate_id_and_species(id_and_raw_data, species_df):
    id_and_species=pd.concat([id_and_raw_data, species_df], axis=1)
    id_and_species.columns = id_and_species.columns.droplevel(1) # supprimer la ligne header translation 
    return id_and_species
```

### Extract planted species

```python
def get_planted_species(df):
    index_begin_plantation=dataframe_total_cleaned.columns.get_loc(('PLANTATION', 'Plantation or Delivery date', "Translation", "Date_of_Planting(1)"))
    index_end_plantation=dataframe_total_cleaned.columns.get_loc(('PLANTATION', 'Total planted trees', "Translation", ""))
    df_plantation = dataframe_total_cleaned.iloc[:, index_begin_plantation+1:index_end_plantation]
    df_plantation.columns = df_plantation.columns.droplevel(0)
    return df_plantation
```

```python
planted_species=get_planted_species(dataframe_total_cleaned)
planted_species.head()
```

```python
planted_species_with_id=concatenate_id_and_species(first_three_columns,planted_species)
planted_species_with_id.head()
```

```python
def get_m1_species(df):
    index_begin_m1=dataframe_total_cleaned.columns.get_loc(('Monitoring 1', 'Technician name', "Translation", "Technician_name_parcel(1)"))
    index_end_m1=dataframe_total_cleaned.columns.get_loc(('Monitoring 1', 'Total replanted trees at mon 1', "Translation", ""))
    df_m1_species = dataframe_total_cleaned.iloc[:, index_begin_m1+1:index_end_m1]
    df_m1_species.columns = df_m1_species.columns.droplevel(0) # get every species from monitoring m1

    df_replanted_m1 = df_m1_species.iloc[:, :int(len(df_m1_species.columns)/2)] # get replanted m1 species by splitting the dataframe by half
    df_living_m1 = df_m1_species.iloc[:, int(len(df_m1_species.columns)/2):]
    return df_replanted_m1,df_living_m1
```

## Write to excel

```python
# https://stackoverflow.com/questions/70617982/how-to-set-column-headers-to-the-first-row-in-pandas-dataframe
# https://stackoverflow.com/questions/31328861/replacing-header-with-top-row

def make_second_header_row_as_first_data_row (df):
    """ To simplify save to excel, we transform the second level of header row as a data row.
    Because pd.to_excel does not support MultiIndex dataframes (with multiple header levels).
    """
    flattened_headers_df = df.T.reset_index().T.reset_index(drop=True)
    new_header = flattened_headers_df.iloc[0] #grab the first row for the header
    flattened_headers_df = flattened_headers_df[1:] #take the data less the header row
    flattened_headers_df.columns = new_header #set the header row as the df header
    return flattened_headers_df
```

```python
output_file=r"File_with_splitted_sheets.xlsx"

planted_species_with_id_excel = make_second_header_row_as_first_data_row (planted_species_with_id)
replanted_m1_with_id_excel = make_second_header_row_as_first_data_row (replanted_m1_with_id)
living_m1_with_id_excel = make_second_header_row_as_first_data_row (living_m1_with_id)
replanted_m2_with_id_excel = make_second_header_row_as_first_data_row (replanted_m2_with_id)
living_m2_with_id_excel = make_second_header_row_as_first_data_row (living_m2_with_id)

with pd.ExcelWriter(output_file) as writer:
    planted_species_with_id_excel.to_excel(writer, sheet_name='Colonne_Plantation', startrow=2, index=False)
    replanted_m1_with_id_excel.to_excel(writer, sheet_name='Niveau1_M1', startrow=2, index=False)
    living_m1_with_id_excel.to_excel(writer, sheet_name='Niveau2_M1', startrow=2, index=False)
    replanted_m2_with_id_excel.to_excel(writer, sheet_name='Niveau1_M2', startrow=2, index=False)
    living_m2_with_id_excel.to_excel(writer, sheet_name='Niveau2_M2', startrow=2, index=False)
```
